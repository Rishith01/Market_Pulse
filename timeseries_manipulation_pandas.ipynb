```python
import yfinance as yf
import matplotlib.pyplot as plt

data = yf.download(tickers = "MSFT", period = "3mo")
data.head()
```

```python
# Compute 15-day Simple Moving Average (SMA15)
data['SMA15'] = data['Adj Close'].rolling(window=15).mean()

# Filter columns for plotting
plot_data = data[['Adj Close', 'SMA15']]

# Plot the data
plot_data.plot(figsize=(12, 3), title="SMA - 15 days")
plt.show()
```
